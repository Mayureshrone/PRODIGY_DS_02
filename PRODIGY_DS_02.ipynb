```python
plt.figure(figsize=(10, 6))
sns.histplot(data["Subscribers (millions)"], kde=True)
plt.title("Distribution of Subscribers (millions)")
plt.xlabel("Subscribers (millions)")
plt.ylabel("Number of Channels")
plt.show()
```

```python
top_channels = data.nlargest(5, "Subscribers (millions)")

plt.figure(figsize=(10, 6))
sns.barplot(x="Subscribers (millions)", y="Name", data=top_channels, palette="viridis")
plt.title("Top 5 Most Subscribed YouTube Channels")
plt.xlabel("Subscribers (millions)")
plt.ylabel("Channel Name")
plt.show()
```

```python
category_counts = data['Category'].value_counts()

plt.figure(figsize=(10, 6))
ax = category_counts.plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.title('Distribution of Channels by Category')
plt.xlabel('Number of Channels')
plt.ylabel('Category')

# Add labels to the bars
for container in ax.containers:
    ax.bar_label(container)

plt.gca().spines[['top', 'right',]].set_visible(False)
plt.show()
```

```python
language_counts = data['Primary language'].value_counts()

plt.figure(figsize=(10, 6))
ax = language_counts.plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.title('Distribution of Channels by Primary Language')
plt.xlabel('Number of Channels')
plt.ylabel('Primary Language')

# Add labels to the bars
for container in ax.containers:
    ax.bar_label(container)

plt.gca().spines[['top', 'right',]].set_visible(False)
plt.show()
```
